```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('../../../data/mean-daily-temperature-fisher-river.csv',
                 parse_dates=True)
df.rename(columns={'Mean temparature': 'MeanTemp'}, inplace=True)
df.index = df.Date
df.drop('Date', axis=1, inplace=True)
df.head()
```

```python
weekly_ma = df.MeanTemp.rolling(7).mean()
monthly_ma = df.MeanTemp.rolling(30).mean()
weekly_ma_2stride = df.MeanTemp.rolling(7).mean()[::2]
monthly_ma_2stride = df.MeanTemp.rolling(30).mean()[::2]
```

```python
plt.plot(weekly_ma);
```

```python
plt.plot(monthly_ma);
```

```python
plt.plot(weekly_ma_2stride)
plt.plot(monthly_ma_2stride);
```
