```python
from random import random

import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AR
```

# AR

```python
data = [x + random() for x in range(1, 100)]
model = AR(data)
model_fit = model.fit()
yhat = model_fit.predict(len(data), len(data))
yhat
```
